```python
def set_figure_pixel_size (width, height, ppi):
    plt.figure(figsize=(width/float(ppi), height/float(ppi)), dpi=ppi)
```

```python
n = len(data_2020['변환이평'])
g = {}
```

```python
route = "/content/drive/Shareddrives/공모공모/흑백_9696_test/2020"
```

```python
for i in range(5,n-3): 
  g[i] = data_2020['변환이평'][i-5:i]
  diff = (data_2020['변환이평'][i+3] - data_2020['변환이평'][i]) / data_2020['변환이평'][i]
  if diff < -0.015:
    file_route = route + "/-1.5미만"
  elif diff < 0.015:
    file_route = route + "/-1.5이상1.5미만"
  else:
    file_route = route + "/1.5이상"
  set_figure_pixel_size(96, 96, 72)
  plt.plot(g[i], color='black')
  plt.axis('off')  
  plt.savefig(file_route+ f"/img_{i-5}", dpi=72)
  plt.clf()
```
